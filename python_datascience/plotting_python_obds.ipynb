#### Matplotlib

- Basic plotting package 
- Very customisable - every single line/pixel 
- Need alot of code to write a simple plot

```python
# Make x & y variables using numpy 
x = np.arange(0, 5, 0.1)
y = np.sin(x)

print('This is x:')
print(x)

print()

print('This is y:')
print(y)
```

### Matplot lib object orientated interface

```python

fig = plt.figure()   # create figure object - contain axes, graphics, text
ax = plt.axes()      # create axes - a single pane

ax.plot(x, y) # create continuous line plot

ax.set_xlim(0,10)
ax.set_ylim(-1.5,1.5)
ax.set_title('title of plot')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
plt.show()
```

### Matplot lib Pyplot interface

```python
plt.plot(x, y)
plt.xlim(0,10)
plt.ylim(-1.5,1.5)
plt.title('title of plt')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()
```

#### Scatterplot

```python
tips = sns.load_dataset("tips")
display(tips)

# set color to Yes/No depending on value in 'smoker column' 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            data=tips)

# as above plus shape of points for time column
sns.relplot(x="total_bill",
            y="tip",
            hue="smoker",
            style="time",
            data=tips);
```

```python

# col = Time -> create 2 plots (facets) split by time column 
sns.relplot(x="total_bill", 
            y="tip", 
            hue="smoker", 
            col="time", 
            data=tips)
```

#### Pairs plot

```python
iris = sns.load_dataset("iris")
display(iris.head())

# plot pairs plot
# assign plot to object 'g' - sets up grid
g = sns.PairGrid(iris, hue="species")

# plot histogram on diagonal
g.map_diag(plt.hist) 
# plot scatter plots in the offdiagonals
g.map_offdiag(plt.scatter)
# add legend to plot
g.add_legend();
```

### Lineplot example

```python
fmri = sns.load_dataset("fmri")
display(fmri.head())
sns.relplot(x="timepoint",
            y="signal", 
            hue="event", 
            kind="line", 
            data=fmri)

sns.relplot(x="timepoint", 
            y="signal", 
            hue="region", 
            style="event",
            dashes=False, 
            markers=True, 
            kind="line", 
            data=fmri);
```

### Facets

```python
sns.relplot(x="timepoint",
            y="signal",
            hue="subject",
            col="region",
            row="event",
            height=3,
            kind="line",
            estimator=None,
            data=fmri)
```

### Plot Categorical Data

```python
# create violin plots
g = sns.catplot(x="day", 
                y="total_bill", 
                kind="violin",
                inner=None,
                data=tips)

# add individual datapoints -> provide ax argument
sns.swarmplot(x="day",
              y="total_bill",
              color="k",
              size=3,
              data=tips,
              ax=g.ax) 
```

Load read count dataset from week 2 rnaseq pipeline (output from featurecounts)

```python
df=pd.read_csv('/ifs/obds-training/jan21/exercises/rnaseq/room1/counts/countdata.tsv', sep='\t', comment='#')
df
```

Rename columns and remove unwanted columns (Chr, Start, End, Strand, Length) 


```python
counts= df.drop(columns=['Chr','Start', 'End', 'Strand','Length'])
counts 
counts.columns= counts.columns.str.replace('sam/', '').str.replace('.bam', '')

#doing it the df.rename way 
counts.rename(columns={'ERR1755082': 'cd4_r1', 'ERR1755083': 'cd4_r2', 'ERR1755084': 'cd4_r3', 'ERR1755085': 'cd8_r1', 'ERR1755086': 'cd8_r2', 'ERR1755087': 'cd8_r3'}, inplace= True)
counts



```

Filter out genes with no counts


```python
counts['gene_count']= counts.sum(axis=1)
counts
```

```python
counts=counts.set_index(['Geneid'])
counts

filtered_counts = counts[counts['gene_count'] >0].drop(columns='gene_count')
```

```python
filtered_counts
```

Plot scatter plot of correlation of replicates (pairs plot) (you will need to google how to show plots in jupyter notebook)


```python
g = sns.PairGrid(np.log10(filtered_counts+1))
g.map_diag(plt.hist, bins=100)
g.map_offdiag(plt.scatter)
g.add_legend()
```

Tidy data


```python
filtered_counts.reset_index(inplace=True)
filtered_counts
```

```python
long_df=filtered_counts.melt(id_vars=['Geneid'], var_name='sample', value_name='counts')
long_df
```

Plot histogram and density plot of read counts across all samples


```python
sns.histplot(np.log2(long_df['counts']+1), kde=True, bins=20)
```

Plot a violin plot of read counts per sample


```python
sns.catplot(data=long_df, x='sample', y= np.log2(long_df['counts']+1), kind='violin')
```

Identify gene with highest expression in each sample


```python
filtered_counts_gene_index = filtered_counts.set_index('Geneid')
maxValueIndexObj = filtered_counts_gene_index.idxmax()
maxValueIndexObj

#to check that those genes are the ones with the highest value 
filtered_counts[filtered_counts['Geneid'].isin(['l7Rn6', 'Eef1a1', 'Actb'])]
```

Normalise read counts to total reads per sample total reads & convert to 'counts per million (CPM)'


```python
filtered_counts_sums=filtered_counts_gene_index.sum(axis=0)
filtered_counts_cpm=filtered_counts_gene_index.div(filtered_counts_sums, axis=1) * 1e6
filtered_counts_cpm
```

Log transform normalised read counts


```python
normalised_counts=np.log2(filtered_counts_cpm + 1)
normalised_counts
```

Create new dataframe of 100 genes with highest average variance in log cpm across conditions


```python
normalised_counts['variance']= normalised_counts.var(axis=1)
highest_variance_top_100 = normalised_counts.sort_values(by=['variance'], ascending=False).head(100)
highest_variance_top_100
```

Plot the of expression of the top 5 most variable genes (log cpm) across all samples (line plot)


```python
top_5 = normalised_counts.sort_values(by=['variance'], ascending=False).head(5)
top_5.reset_index(inplace=True)
top_5_index=top_5.drop('variance', axis=1)
top_5_index

top_5_melt=top_5_index.melt(id_vars=['Geneid'], var_name='sample', value_name='counts')


top_5_melt


sns.relplot(data=top_5_melt, x="sample", y="counts", hue="Geneid", kind='line')
```

Plot a heatmap for top 100 most variable genes (log cpm) across all samples


```python
top_100_index=highest_variance_top_100.drop('variance', axis=1)
g=sns.clustermap(top_100_index)
```
