```python
#Subset the rows that have reads_per_base values greater than 0.05 and a chromosome length under 100 Mb
idxdf[(idxdf['reads_per_base'] > 0.05) & (idxdf['length_Mbs'] < 100)]

#Delete the reads_per_base column
idxdf = idxdf.drop('reads_per_base', axis=1)

idxdf


```

```python
#read in ensembl gtf files to create gtf
gtf = pd.read_table('/ifs/obds-training/jan21/exercises/pandas/Mus_musculus.GRCm38.102.chr.gtf.gz', sep='\t', header=None, comment='#')



```

```python
#Assign appropriate column names https://www.ensembl.org/info/website/upload/gff.html
gtf.columns = ['seq_name', 'source', 'feature','start', 'end', 'score', 'strand', 'frame', 'attribute']
print(gtf) 

#Find the unique values for 'source' column
print(gtf['source'].unique())

#Subset all values from 'ensembl_havana' and assign to new df called 'ens_df'
ens_df = gtf[gtf.source == 'ensembl_havana']

#What is the size of ens_df?
ens_df.info
ens_df.head()

#Group ens_df by 'feature' column & apply the 'count' function to get the number of rows in each feature
ens_df.groupby('feature').count()

#Create a new df called 'genes' by subsetting all rows with 'gene' as a feature
gene_df= ens_df[ens_df.feature == 'gene'].copy()
display(gene_df)

#Create a new column called 'gene_length' and calculate the length of each gene
gene_df['gene_length']= gene_df['end']- gene_df['start']
display(gene_df)

#Calculate the mean and median of gene lengths
display(gene_df.mean(axis=0))
display(gene_df.median(axis=0))
display(np.median(gene_df['gene_length']))
display(np.mean(gene_df['gene_length']))
display(gene_df['gene_length'].describe())


#Find the top 10 longest genes
gene_df.sort_values(by=['gene_length'], ascending=False).head(10)

```

```python
#Find the lowest 20 shortest genes
gene_df.sort_values(by=['gene_length'], ascending=True).head(20)

#Split attributes column on ';' and create new dataframe, merge this onto the ens_gtf dataframe using pandas merge 
attributes_df=gene_df['attribute'].str.split(';', expand=True)
display(attributes_df)
display(gene_df)

genes_full_attributes = pd.concat([gene_df, attributes_df], axis=1)

display(genes_full_attributes)




```
